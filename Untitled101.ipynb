```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.metrics import (
    accuracy_score, confusion_matrix, classification_report,
    roc_curve, auc
)
from sklearn.ensemble import GradientBoostingClassifier
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import learning_curve

warnings.filterwarnings("ignore")

# -------------------------------
# 1. Load Training Dataset (z310ab.csv)
# -------------------------------
train_path = "c:/z310ab.csv"
df = pd.read_csv(train_path)

X = df.drop("DiagnosisType", axis=1)
y = df["DiagnosisType"]

# -------------------------------
# 2. Preprocessing
# -------------------------------
imputer = IterativeImputer()
scaler = StandardScaler()

X_imputed = imputer.fit_transform(X)
X_scaled = scaler.fit_transform(X_imputed)

# -------------------------------
# 3. Apply SMOTE to Full Dataset
# -------------------------------
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

print(f"\nâœ… Original samples: {len(X)}, After SMOTE: {len(X_resampled)}")

# -------------------------------
# 4. Train Model on Full Resampled Data
# -------------------------------
model = GradientBoostingClassifier(random_state=42)
model.fit(X_resampled, y_resampled)

# -------------------------------
# 5. Evaluate on Full Training Data (for insight)
# -------------------------------
y_pred_train = model.predict(X_resampled)
acc_train = accuracy_score(y_resampled, y_pred_train)

print(f"\n=== Full Resampled Training Data ===")
print(f"Accuracy: {acc_train:.4f}")
print(classification_report(y_resampled, y_pred_train))

# Confusion Matrix
cm_train = confusion_matrix(y_resampled, y_pred_train)
sns.heatmap(cm_train, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Full Resampled Training Data")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# ROC Curve for Training Data
y_train_bin = label_binarize(y_resampled, classes=np.unique(y))
y_train_proba = model.predict_proba(X_resampled)
fpr_train, tpr_train, _ = roc_curve(y_train_bin.ravel(), y_train_proba.ravel())
roc_auc_train = auc(fpr_train, tpr_train)

plt.plot(fpr_train, tpr_train, color="blue", label=f"Training AUC = {roc_auc_train:.4f}")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Full Resampled Training Data")
plt.legend()
plt.show()

# -------------------------------
# 6. Evaluate on External Dataset
# -------------------------------
external_path = "c:/externalDiabetes_predictions.csv"
external_df = pd.read_csv(external_path)

if "DiagnosisType" in external_df.columns:
    y_external_true = external_df["DiagnosisType"]
    X_external = external_df.drop("DiagnosisType", axis=1)
else:
    raise ValueError("External dataset must contain 'DiagnosisType' column.")

X_ext_imputed = imputer.transform(X_external)
X_ext_scaled = scaler.transform(X_ext_imputed)
smote = SMOTE(random_state=42)
X_ext_scaled, y_external_true = smote.fit_resample(X_ext_scaled, y_external_true)

y_ext_pred = model.predict(X_ext_scaled)
acc_ext = accuracy_score(y_external_true, y_ext_pred)

print(f"\n=== External Dataset ===")
print(f"Accuracy: {acc_ext:.4f}")
print(classification_report(y_external_true, y_ext_pred))
cm_ext = confusion_matrix(y_external_true, y_ext_pred)
sns.heatmap(cm_ext, annot=True, fmt="d", cmap="Oranges")
plt.title("Confusion Matrix - External Dataset")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# ROC Curve (External Dataset)
y_ext_bin = label_binarize(y_external_true, classes=np.unique(y))
y_ext_proba = model.predict_proba(X_ext_scaled)
fpr_ext, tpr_ext, _ = roc_curve(y_ext_bin.ravel(), y_ext_proba.ravel())
roc_auc_ext = auc(fpr_ext, tpr_ext)

plt.plot(fpr_ext, tpr_ext, color="red", label=f"External AUC = {roc_auc_ext:.4f}")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - External Dataset")
plt.legend()
plt.show()

# -------------------------------
# 7. Combined ROC Curve Plot
# -------------------------------
plt.figure(figsize=(7,6))
plt.plot(fpr_train, tpr_train, label=f"Training AUC = {roc_auc_train:.4f}", color="blue")
plt.plot(fpr_ext, tpr_ext, label=f"External AUC = {roc_auc_ext:.4f}", color="red")
plt.plot([0, 1], [0, 1], 'k--', lw=1)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison (Training vs External)")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------
# 8. Accuracy & AUC Comparison Bar Chart
# -------------------------------
metrics_df = pd.DataFrame({
    "Dataset": ["Full Resampled Training", "External Dataset"],
    "Accuracy": [acc_train, acc_ext],
    "AUC": [roc_auc_train, roc_auc_ext]
})

fig, axes = plt.subplots(1, 2, figsize=(10,5))

sns.barplot(x="Dataset", y="Accuracy", data=metrics_df, ax=axes[0], palette="Blues")
axes[0].set_ylim(0, 1)
axes[0].set_title("Accuracy Comparison")

sns.barplot(x="Dataset", y="AUC", data=metrics_df, ax=axes[1], palette="Oranges")
axes[1].set_ylim(0, 1)
axes[1].set_title("AUC Comparison")

plt.tight_layout()
plt.show()

# -------------------------------
# 9. Learning Curve on Full Resampled Data
# -------------------------------
train_sizes, train_scores, test_scores = learning_curve(
    model, X_resampled, y_resampled,
    cv=5, scoring="accuracy",
    train_sizes=np.linspace(0.1, 1.0, 5), n_jobs=-1
)

train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

plt.plot(train_sizes, train_mean, 'o-', label="Training Accuracy")
plt.plot(train_sizes, test_mean, 'o-', label="Validation Accuracy")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.title("Learning Curve (Full Resampled Data)")
plt.legend()
plt.grid(True)
plt.show()

# -------------------------------
# 10. Final Performance Statement in Arabic
# -------------------------------
if acc_ext >= 0.80:
    print(f"\nâœ…  The prddiction is External Dataset ExcellentØ§ (Accuracy = {acc_ext:.2%})")
elif acc_ext >= 0.60:
    print(f"\nâš ï¸ The prddiction is External Dataset Moderate (Accuracy = {acc_ext:.2%})")
else:
    print(f"\nâŒ The prddiction is External Dataset Weak (Accuracy = {acc_ext:.2%})")
```
