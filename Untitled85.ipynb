```python

import os
import warnings
    
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support
)

from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from pyswarm import pso
from itertools import cycle

warnings.filterwarnings('ignore')
print ("################################################### Random Forest")
# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/z310ab.csv")
target_column = "DiagnosisType"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)
features_df = pd.DataFrame(features_imputed, columns=features.columns)

# Standardize
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_df)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", y.value_counts())

# Apply KMeansSMOTE followed by SMOTE
kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", pd.Series(y_res).value_counts())

# -------------------------------
# 3. Train Random Forest
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_test = rf_model.predict(X_test)
y_score_test = rf_model.predict_proba(X_test)

# Evaluation
print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation (optional)
# -------------------------------
try:
    explainer = shap.Explainer(rf_model, X_train)
    shap_values = explainer(X_train[:100])
    shap.summary_plot(shap_values, pd.DataFrame(X_train[:100], columns=X.columns))
except Exception as e:
    print(f"SHAP explanation error: {e}")

# -------------------------------
## Optional: Confusion Matrix on Full Dataset
# -------------------------------
y_full_pred = rf_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# Optional: Class Distribution in Test Set
# -------------------------------
print("\nTest Set Class Distribution:")
print(y_test.value_counts())

# -------------------------------
# Optional: Confusion Matrix on Full Dataset
# -------------------------------
y_full_pred = rf_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(10, 8))
colors = cycle(['green', 'orange', 'red', 'blue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass Random Forest ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    rf_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Random Forest")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. PSO Hyperparameter Tuning
# -------------------------------
def rf_loss_function(params):
    n_estimators, max_depth = int(params[0]), int(params[1])
    clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    return 1 - accuracy_score(y_test, preds)

lb = [50, 2]
ub = [300, 20]
best_params, _ = pso(rf_loss_function, lb, ub, swarmsize=10, maxiter=5)
print("\nBest Hyperparameters (via PSO):", best_params)

# -------------------------------
# 9. Evaluate Optimized Model
# -------------------------------
opt_rf = RandomForestClassifier(n_estimators=int(best_params[0]), max_depth=int(best_params[1]), random_state=42)
opt_rf.fit(X_train, y_train)

opt_preds = opt_rf.predict(X_test)
opt_proba = opt_rf.predict_proba(X_test)

opt_acc = accuracy_score(y_test, opt_preds)
opt_auc = roc_auc_score(y_test_bin, opt_proba, multi_class='ovr', average='weighted')
opt_f1 = precision_recall_fscore_support(y_test, opt_preds, average='weighted')[2]

comparison = pd.DataFrame({
    "Metric": ["Accuracy", "F1 Score", "AUC"],
    "Before PSO": [
        accuracy_score(y_test, y_pred_test),
        precision_recall_fscore_support(y_test, y_pred_test, average='weighted')[2],
        auc_score
    ],
    "After PSO": [opt_acc, opt_f1, opt_auc]
})

print("\nComparison: Before vs After PSO")
print(comparison.to_string(index=False))

# -------------------------------
# 10. Predict on External Dataset
# -------------------------------
external_path = "c:/z310ab01.csv"
external_raw = pd.read_csv(external_path)

# Drop target if present
if target_column in external_raw.columns:
    external_raw = external_raw.drop(columns=[target_column])

external_raw = external_raw[X.columns]

external_imputed = imputer.transform(external_raw)
external_scaled = scaler.transform(external_imputed)

external_preds = opt_rf.predict(external_scaled)
external_proba = opt_rf.predict_proba(external_scaled)

external_results = external_raw.copy()
external_results["Predicted"] = external_preds
external_results["Diagnosis Label"] = external_results["Predicted"].map({
    0: "Normal", 1: "Type1", 2: "Type2", 3: "Gestational"
})

output_path = "C:/Users/عبير/external_predictions.csv"
external_results.to_csv(output_path, index=False)
print(f"\nPredictions saved to {output_path}")
from sklearn.metrics import mean_squared_error, r2_score

# 1. Overall Accuracy (already printed)
overall_accuracy = accuracy_score(y_test, y_pred_test)

# 2. Overall Precision, Recall
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted'
)

# 3. Training Score
training_score = rf_model.score(X_train, y_train)

# 4. Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred_test)

# 5. R^2 Score
r2 = r2_score(y_test, y_pred_test)

# Display Results
print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {overall_accuracy:.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")
print ("################################################### Logistic Regression")
# diabetes_pipeline.py

import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support
)

from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from pyswarm import pso
from itertools import cycle

warnings.filterwarnings('ignore')
print("################################################### Logistic Regression")

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/z310ab.csv")
target_column = "DiagnosisType"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)
features_df = pd.DataFrame(features_imputed, columns=features.columns)

# Standardize
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_df)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", y.value_counts())

# Apply KMeansSMOTE followed by SMOTE
kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", pd.Series(y_res).value_counts())

# -------------------------------
# 3. Train Logistic Regression
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

lr_model = LogisticRegression(max_iter=1000, random_state=42, multi_class='ovr')
lr_model.fit(X_train, y_train)

y_pred_test = lr_model.predict(X_test)
y_score_test = lr_model.predict_proba(X_test)

# Evaluation
print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation (optional)
# -------------------------------
try:
    explainer = shap.Explainer(lr_model, X_train)
    shap_values = explainer(X_train[:100])
    shap.summary_plot(shap_values, pd.DataFrame(X_train[:100], columns=X.columns))
except Exception as e:
    print(f"SHAP explanation error: {e}")

# -------------------------------
# 5. Confusion Matrix (Test Set)
# Optional: Confusion Matrix on Full Dataset
# -------------------------------
y_full_pred = rf_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# Optional: Class Distribution in Test Set
# -------------------------------
print("\nTest Set Class Distribution:")
print(y_test.value_counts())

# -------------------------------
# Optional: Confusion Matrix on Full Dataset
# -------------------------------
y_full_pred = lr_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(10, 8))
colors = cycle(['green', 'orange', 'red', 'blue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass Logistic Regression ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    lr_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Logistic Regression")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. PSO Hyperparameter Tuning (example: tuning C parameter)
# -------------------------------
def lr_loss_function(params):
    C = params[0]
    clf = LogisticRegression(C=C, max_iter=1000, random_state=42, multi_class='ovr')
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    return 1 - accuracy_score(y_test, preds)

lb = [0.01]
ub = [10]
best_params, _ = pso(lr_loss_function, lb, ub, swarmsize=10, maxiter=5)
print("\nBest Hyperparameters (via PSO):", best_params)

# -------------------------------
# 9. Evaluate Optimized Model
# -------------------------------
opt_lr = LogisticRegression(C=best_params[0], max_iter=1000, random_state=42, multi_class='ovr')
opt_lr.fit(X_train, y_train)

opt_preds = opt_lr.predict(X_test)
opt_proba = opt_lr.predict_proba(X_test)

opt_acc = accuracy_score(y_test, opt_preds)
opt_auc = roc_auc_score(y_test_bin, opt_proba, multi_class='ovr', average='weighted')
opt_f1 = precision_recall_fscore_support(y_test, opt_preds, average='weighted')[2]

comparison = pd.DataFrame({
    "Metric": ["Accuracy", "F1 Score", "AUC"],
    "Before PSO": [
        accuracy_score(y_test, y_pred_test),
        precision_recall_fscore_support(y_test, y_pred_test, average='weighted')[2],
        auc_score
    ],
    "After PSO": [opt_acc, opt_f1, opt_auc]
})

print("\nComparison: Before vs After PSO")
print(comparison.to_string(index=False))

# -------------------------------
# 10. Predict on External Dataset
# -------------------------------
external_path = "c:/z310ab01.csv"
external_raw = pd.read_csv(external_path)

# Drop target if present
if target_column in external_raw.columns:
    external_raw = external_raw.drop(columns=[target_column])

external_raw = external_raw[X.columns]

external_imputed = imputer.transform(external_raw)
external_scaled = scaler.transform(external_imputed)

external_preds = opt_lr.predict(external_scaled)
external_proba = opt_lr.predict_proba(external_scaled)

external_results = external_raw.copy()
external_results["Predicted"] = external_preds
external_results["Diagnosis Label"] = external_results["Predicted"].map({
    0: "Normal", 1: "Type1", 2: "Type2", 3: "Gestational"
})

output_path = "C:/Users/عبير/external_predictions.csv"
external_results.to_csv(output_path, index=False)
print(f"\nPredictions saved to {output_path}")

from sklearn.metrics import mean_squared_error, r2_score

# 1. Overall Accuracy (already printed)
overall_accuracy = accuracy_score(y_test, y_pred_test)

# 2. Overall Precision, Recall
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted'
)

# 3. Training Score
training_score = lr_model.score(X_train, y_train)

# 4. Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred_test)

# 5. R^2 Score
r2 = r2_score(y_test, y_pred_test)

# Display Results
print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {overall_accuracy:.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")

print ("################################################### Gausian Bayes")
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.naive_bayes import GaussianNB
from imblearn.over_sampling import SMOTE, KMeansSMOTE
from itertools import cycle

import warnings
warnings.filterwarnings("ignore")

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/z310ab.csv")
target_column = "DiagnosisType"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", y.value_counts())

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", pd.Series(y_res).value_counts())

# -------------------------------
# 3. Train Gaussian Naive Bayes
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

y_pred_test = nb_model.predict(X_test)
y_score_test = nb_model.predict_proba(X_test)

# -------------------------------
# 4. Evaluation
# -------------------------------
print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# Compute custom metrics
accuracy = accuracy_score(y_test, y_pred_test)
report = classification_report(y_test, y_pred_test, output_dict=True)
precision = report['weighted avg']['precision']
recall = report['weighted avg']['recall']
train_score = nb_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print(f"\nOverall Accuracy: {accuracy:.4f}")
print(f"Overall Precision: {precision:.4f}")
print(f"Overall Recall: {recall:.4f}")
print(f"Training Score: {train_score:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R2 Score: {r2:.4f}")

# -------------------------------

# Optional: Confusion Matrix on Full Dataset
# -------------------------------
y_full_pred = nb_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()
# -------------------------------
# 6. ROC Curve
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]

fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(10, 8))
colors = cycle(['green', 'orange', 'red', 'blue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass GaussianNB ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    nb_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Gaussian Naive Bayes")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Predict on External Dataset
# -------------------------------
external_path = "c:/z310ab01.csv"
external_raw = pd.read_csv(external_path)

if target_column in external_raw.columns:
    external_raw = external_raw.drop(columns=[target_column])

external_raw = external_raw[X.columns]

external_imputed = imputer.transform(external_raw)
external_scaled = scaler.transform(external_imputed)

external_preds = nb_model.predict(external_scaled)
external_proba = nb_model.predict_proba(external_scaled)

external_results = external_raw.copy()
external_results["Predicted"] = external_preds
external_results["Diagnosis Label"] = external_results["Predicted"].map({
    0: "Normal", 1: "Type1", 2: "Type2", 3: "Gestational"
})

output_path = "C:/Users/عبير/external_predictions_nb.csv"
external_results.to_csv(output_path, index=False)
print(f"\nPredictions saved to {output_path}")

print("#########################SVM")

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE, KMeansSMOTE
from itertools import cycle

import warnings
warnings.filterwarnings("ignore")

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/z310ab.csv")
target_column = "DiagnosisType"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", y.value_counts())

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", pd.Series(y_res).value_counts())

# -------------------------------
# 3. Train SVM Classifier
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

svm_model = SVC(probability=True, kernel='rbf', C=1.0, gamma='scale', random_state=42)
svm_model.fit(X_train, y_train)

y_pred_test = svm_model.predict(X_test)
y_score_test = svm_model.predict_proba(X_test)

# -------------------------------
# 4. Evaluation
# -------------------------------
print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# Compute custom metrics
accuracy = accuracy_score(y_test, y_pred_test)
report = classification_report(y_test, y_pred_test, output_dict=True)
precision = report['weighted avg']['precision']
recall = report['weighted avg']['recall']
train_score = svm_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print(f"\nOverall Accuracy: {accuracy:.4f}")
print(f"Overall Precision: {precision:.4f}")
print(f"Overall Recall: {recall:.4f}")
print(f"Training Score: {train_score:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R2 Score: {r2:.4f}")

# -------------------------------
# 5. Confusion Matrix
# -------------------------------

# Optional: Confusion Matrix on Full Dataset
# -------------------------------
y_full_pred = svm_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()
# -------------------------------
# 6. ROC Curve
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]

fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(10, 8))
colors = cycle(['green', 'orange', 'red', 'blue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - SVM')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    svm_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=1  # Use n_jobs=1 to avoid memory error
)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - SVM")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Predict on External Dataset
# -------------------------------
external_path = "c:/z310ab01.csv"
external_raw = pd.read_csv(external_path)

if target_column in external_raw.columns:
    external_raw = external_raw.drop(columns=[target_column])

external_raw = external_raw[X.columns]

external_imputed = imputer.transform(external_raw)
external_scaled = scaler.transform(external_imputed)

external_preds = svm_model.predict(external_scaled)
external_proba = svm_model.predict_proba(external_scaled)

external_results = external_raw.copy()
external_results["Predicted"] = external_preds
external_results["Diagnosis Label"] = external_results["Predicted"].map({
    0: "Normal", 1: "Type1", 2: "Type2", 3: "Gestational"
})

output_path = "C:/Users/عبير/external_predictions_svm.csv"
external_results.to_csv(output_path, index=False)
print(f"\nPredictions saved to {output_path}")
print("################ Adaboost")
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, mean_squared_error, r2_score
)

from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
from itertools import cycle

import warnings
warnings.filterwarnings("ignore")

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/z310ab.csv")
target_column = "DiagnosisType"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", y.value_counts())

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", pd.Series(y_res).value_counts())

# -------------------------------
# 3. Train AdaBoost Classifier
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

adaboost_model = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators=50,
    learning_rate=1.0,
    algorithm='SAMME',  # ✅ change from 'SAMME.R'
    random_state=42
)

adaboost_model.fit(X_train, y_train)

y_pred_test = adaboost_model.predict(X_test)
y_score_test = adaboost_model.predict_proba(X_test)

# -------------------------------
# 4. Evaluation
# -------------------------------
print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

accuracy = accuracy_score(y_test, y_pred_test)
report = classification_report(y_test, y_pred_test, output_dict=True)
precision = report['weighted avg']['precision']
recall = report['weighted avg']['recall']
train_score = adaboost_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print(f"\nOverall Accuracy: {accuracy:.4f}")
print(f"Overall Precision: {precision:.4f}")
print(f"Overall Recall: {recall:.4f}")
print(f"Training Score: {train_score:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R2 Score: {r2:.4f}")

# -------------------------------
# 5. Confusion Matrix
# -------------------------------

# Optional: Confusion Matrix on Full Dataset
# -------------------------------
y_full_pred = adaboost_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()
# -------------------------------
# 6. ROC Curve
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(10, 8))
colors = cycle(['green', 'orange', 'red', 'blue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - AdaBoost')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    adaboost_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=1  # Avoid parallel memory errors
)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - AdaBoost")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Predict on External Dataset
# -------------------------------
external_path = "c:/z310ab01.csv"
external_raw = pd.read_csv(external_path)

if target_column in external_raw.columns:
    external_raw = external_raw.drop(columns=[target_column])

external_raw = external_raw[X.columns]

external_imputed = imputer.transform(external_raw)
external_scaled = scaler.transform(external_imputed)

external_preds = adaboost_model.predict(external_scaled)
external_proba = adaboost_model.predict_proba(external_scaled)

external_results = external_raw.copy()
external_results["Predicted"] = external_preds
external_results["Diagnosis Label"] = external_results["Predicted"].map({
    0: "Normal", 1: "Type1", 2: "Type2", 3: "Gestational"
})

output_path = "C:/Users/عبير/external_predictions_adaboost.csv"
external_results.to_csv(output_path, index=False)
print(f"\nPredictions saved to {output_path}")


print("########### KNN")
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, mean_squared_error, r2_score
)

from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
from itertools import cycle

import warnings
warnings.filterwarnings("ignore")

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/z310ab.csv")
target_column = "DiagnosisType"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", y.value_counts())

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", pd.Series(y_res).value_counts())

# -------------------------------
# 3. Train K-Nearest Neighbors
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

y_pred_test = knn_model.predict(X_test)
y_score_test = knn_model.predict_proba(X_test)

# -------------------------------
# 4. Evaluation
# -------------------------------
print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

accuracy = accuracy_score(y_test, y_pred_test)
report = classification_report(y_test, y_pred_test, output_dict=True)
precision = report['weighted avg']['precision']
recall = report['weighted avg']['recall']
train_score = knn_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print(f"\nOverall Accuracy: {accuracy:.4f}")
print(f"Overall Precision: {precision:.4f}")
print(f"Overall Recall: {recall:.4f}")
print(f"Training Score: {train_score:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R2 Score: {r2:.4f}")

# -------------------------------
# 5. Confusion Matrix
# -------------------------------

# -------------------------------
y_full_pred = knn_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()
# -------------------------------
# 6. ROC Curve
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]

fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(10, 8))
colors = cycle(['green', 'orange', 'red', 'blue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - KNN')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    knn_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=1
)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - KNN")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Predict on External Dataset
# -------------------------------
external_path = "c:/z310ab01.csv"
external_raw = pd.read_csv(external_path)

if target_column in external_raw.columns:
    external_raw = external_raw.drop(columns=[target_column])

external_raw = external_raw[X.columns]

external_imputed = imputer.transform(external_raw)
external_scaled = scaler.transform(external_imputed)

external_preds = knn_model.predict(external_scaled)
external_proba = knn_model.predict_proba(external_scaled)

external_results = external_raw.copy()
external_results["Predicted"] = external_preds
external_results["Diagnosis Label"] = external_results["Predicted"].map({
    0: "Normal", 1: "Type1", 2: "Type2", 3: "Gestational"
})

output_path = "C:/Users/عبير/external_predictions_knn.csv"
external_results.to_csv(output_path, index=False)
print(f"\nPredictions saved to {output_path}")
print("########### Decision Trees")
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, mean_squared_error, r2_score
)

from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
from itertools import cycle

import warnings
warnings.filterwarnings("ignore")

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/z310ab.csv")
target_column = "DiagnosisType"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", y.value_counts())

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", pd.Series(y_res).value_counts())

# -------------------------------
# 3. Train Decision Tree Classifier
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

dt_model = DecisionTreeClassifier(max_depth=6, random_state=42)
dt_model.fit(X_train, y_train)

y_pred_test = dt_model.predict(X_test)
y_score_test = dt_model.predict_proba(X_test)

# -------------------------------
# 4. Evaluation
# -------------------------------
print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

accuracy = accuracy_score(y_test, y_pred_test)
report = classification_report(y_test, y_pred_test, output_dict=True)
precision = report['weighted avg']['precision']
recall = report['weighted avg']['recall']
train_score = dt_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print(f"\nOverall Accuracy: {accuracy:.4f}")
print(f"Overall Precision: {precision:.4f}")
print(f"Overall Recall: {recall:.4f}")
print(f"Training Score: {train_score:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R2 Score: {r2:.4f}")

# -------------------------------
# 5. Confusion Matrix
# -------------------------------


# Optional: Confusion Matrix on Full Dataset
# -------------------------------
y_full_pred = dt_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()
# -------------------------------
# 6. ROC Curve
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(10, 8))
colors = cycle(['green', 'orange', 'red', 'blue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - Decision Tree')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    dt_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=1
)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Decision Tree")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Predict on External Dataset
# -------------------------------
external_path = "c:/z310ab01.csv"
external_raw = pd.read_csv(external_path)

if target_column in external_raw.columns:
    external_raw = external_raw.drop(columns=[target_column])

external_raw = external_raw[X.columns]

external_imputed = imputer.transform(external_raw)
external_scaled = scaler.transform(external_imputed)

external_preds = dt_model.predict(external_scaled)
external_proba = dt_model.predict_proba(external_scaled)

external_results = external_raw.copy()
external_results["Predicted"] = external_preds
external_results["Diagnosis Label"] = external_results["Predicted"].map({
    0: "Normal", 1: "Type1", 2: "Type2", 3: "Gestational"
})

output_path = "C:/Users/عبير/external_predictions_dt.csv"
external_results.to_csv(output_path, index=False)
print(f"\nPredictions saved to {output_path}")
print("############ Gradient Boosting")
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, mean_squared_error, r2_score
)

from sklearn.ensemble import GradientBoostingClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
from itertools import cycle

import warnings
warnings.filterwarnings("ignore")

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/z310ab.csv")
target_column = "DiagnosisType"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", y.value_counts())

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", pd.Series(y_res).value_counts())

# -------------------------------
# 3. Train Gradient Boosting Classifier
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gb_model.fit(X_train, y_train)

y_pred_test = gb_model.predict(X_test)
y_score_test = gb_model.predict_proba(X_test)

# -------------------------------
# 4. Evaluation
# -------------------------------
print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

accuracy = accuracy_score(y_test, y_pred_test)
report = classification_report(y_test, y_pred_test, output_dict=True)
precision = report['weighted avg']['precision']
recall = report['weighted avg']['recall']
train_score = gb_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print(f"\nOverall Accuracy: {accuracy:.4f}")
print(f"Overall Precision: {precision:.4f}")
print(f"Overall Recall: {recall:.4f}")
print(f"Training Score: {train_score:.4f}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R2 Score: {r2:.4f}")

# -------------------------------
## Optional: Confusion Matrix on Full Dataset
# -------------------------------
y_full_pred = gb_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)

plt.figure(figsize=(6, 5))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(10, 8))
colors = cycle(['green', 'orange', 'red', 'blue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - Gradient Boosting')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    gb_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=1
)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Gradient Boosting")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Predict on External Dataset
# -------------------------------
external_path = "c:/z310ab01.csv"
external_raw = pd.read_csv(external_path)

if target_column in external_raw.columns:
    external_raw = external_raw.drop(columns=[target_column])

external_raw = external_raw[X.columns]

external_imputed = imputer.transform(external_raw)
external_scaled = scaler.transform(external_imputed)

external_preds = gb_model.predict(external_scaled)
external_proba = gb_model.predict_proba(external_scaled)

external_results = external_raw.copy()
external_results["Predicted"] = external_preds
external_results["Diagnosis Label"] = external_results["Predicted"].map({
    0: "Normal", 1: "Type1", 2: "Type2", 3: "Gestational"
})

output_path = "C:/Users/عبير/external_predictions_gb.csv"
external_results.to_csv(output_path, index=False)
print(f"\nPredictions saved to {output_path}")

data_classes = ['Normal', 'Type1', 'Type2', 'Gestational']
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]

fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(10, 8))
colors = cycle(['green', 'orange', 'red', 'blue'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - Gradient Boosting')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    gb_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=1
)

plt.figure(figsize=(8, 6))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Gradient Boosting")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Predict on External Dataset
# -------------------------------
external_path = "c:/z310ab01.csv"
external_raw = pd.read_csv(external_path)

if target_column in external_raw.columns:
    external_raw = external_raw.drop(columns=[target_column])

external_raw = external_raw[X.columns]

external_imputed = imputer.transform(external_raw)
external_scaled = scaler.transform(external_imputed)

external_preds = gb_model.predict(external_scaled)
external_proba = gb_model.predict_proba(external_scaled)

external_results = external_raw.copy()
external_results["Predicted"] = external_preds
external_results["Diagnosis Label"] = external_results["Predicted"].map({
    0: "Normal", 1: "Type1", 2: "Type2", 3: "Gestational"
})

output_path = "C:/Users/عبير/external_predictions_gb.csv"
external_results.to_csv(output_path, index=False)
print(f"\nPredictions saved to {output_path}")
    
```
