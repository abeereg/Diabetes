```python
# diabetes_pipeline.py

import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from pyswarm import pso
from itertools import cycle
from collections import Counter

warnings.filterwarnings('ignore')

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/a.csv")
target_column = "Target"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", Counter(y))

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", Counter(y_res))

# -------------------------------
# 3. Train Random Forest
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_test = rf_model.predict(X_test)
y_score_test = rf_model.predict_proba(X_test)

print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation
# -------------------------------
def explain_model_with_shap(model, X_data, feature_names=None, num_samples=100):
    try:
        explainer = shap.Explainer(model, X_data, check_additivity=False)
        sample = X_data[:num_samples]
        sample_df = pd.DataFrame(sample, columns=feature_names)
        shap_values = explainer(sample)
        shap.summary_plot(shap_values, sample_df)
        return shap_values
    except Exception as e:
        print(f"SHAP explanation error: {e}")
        return None

shap_values = explain_model_with_shap(rf_model, X_train, feature_names=X.columns, num_samples=100)

# -------------------------------
# 5. Confusion Matrix (Full Set)
# -------------------------------
data_classes = ['Type 1','Type 2','Type 3c','Gestational','MODY','LADA','Secondary','Neonatal Mellitus',
                'Wolcott-Rallison Syndrome','Steroid-Induced','Cystic Fibrosis-Related','Wolfram Syndrome','Prediabetic']

y_full_pred = rf_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve (Multiclass)
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(12, 10))
colors = cycle(['green', 'orange', 'red', 'blue', 'purple', 'brown', 'pink', 'gray'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - Random Forest')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    rf_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(20, 20))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Random Forest")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. PSO Hyperparameter Tuning
# -------------------------------
def rf_loss_function(params):
    n_estimators, max_depth = int(params[0]), int(params[1])
    clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=42)
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    return 1 - accuracy_score(y_test, preds)

lb = [50, 2]
ub = [300, 20]
best_params, _ = pso(rf_loss_function, lb, ub, swarmsize=10, maxiter=5)
print("\nBest Hyperparameters (via PSO):", best_params)

# -------------------------------
# 9. Evaluate Optimized Model
# -------------------------------
opt_rf = RandomForestClassifier(n_estimators=int(best_params[0]), max_depth=int(best_params[1]), random_state=42)
opt_rf.fit(X_train, y_train)
opt_preds = opt_rf.predict(X_test)
opt_proba = opt_rf.predict_proba(X_test)

opt_acc = accuracy_score(y_test, opt_preds)
opt_auc = roc_auc_score(y_test_bin, opt_proba, multi_class='ovr', average='weighted')
opt_f1 = precision_recall_fscore_support(y_test, opt_preds, average='weighted')[2]

comparison = pd.DataFrame({
    "Metric": ["Accuracy", "F1 Score", "AUC"],
    "Before PSO": [
        accuracy_score(y_test, y_pred_test),
        precision_recall_fscore_support(y_test, y_pred_test, average='weighted')[2],
        auc_score
    ],
    "After PSO": [opt_acc, opt_f1, opt_auc]
})

print("\nComparison: Before vs After PSO")
print(comparison.to_string(index=False))

# -------------------------------
# 10. Additional Metrics
# -------------------------------
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted')
training_score = rf_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {accuracy_score(y_test, y_pred_test):.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")
################################
# diabetes_pipeline.py
############################################
import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from itertools import cycle
from collections import Counter

warnings.filterwarnings('ignore')

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/a.csv")
target_column = "Target"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", Counter(y))

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", Counter(y_res))

# -------------------------------
# 3. Train Decision Tree Classifier
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

decision_tree_model = DecisionTreeClassifier(
    criterion='gini',      # or 'entropy'
    max_depth=None,        # No depth limit
    random_state=42
)
decision_tree_model.fit(X_train, y_train)

y_pred_test = decision_tree_model.predict(X_test)
y_score_test = decision_tree_model.predict_proba(X_test)

print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation
# -------------------------------
def explain_model_with_shap(model, X_data, feature_names=None, num_samples=100):
    try:
        explainer = shap.TreeExplainer(model)  # TreeExplainer for decision trees
        sample = X_data[:num_samples]
        sample_df = pd.DataFrame(sample, columns=feature_names)
        shap_values = explainer.shap_values(sample)
        shap.summary_plot(shap_values, sample_df)
        return shap_values
    except Exception as e:
        print(f"SHAP explanation error: {e}")
        return None

shap_values = explain_model_with_shap(decision_tree_model, X_train, feature_names=X.columns, num_samples=100)

# -------------------------------
# 5. Confusion Matrix (Full Set)
# -------------------------------
data_classes = ['Type 1','Type 2','Type 3c','Gestational','MODY','LADA','Secondary','Neonatal Mellitus',
                'Wolcott-Rallison Syndrome','Steroid-Induced','Cystic Fibrosis-Related','Wolfram Syndrome','Prediabetic']

y_full_pred = decision_tree_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve (Multiclass)
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(12, 10))
colors = cycle(['green', 'orange', 'red', 'blue', 'purple', 'brown', 'pink', 'gray'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - Decision Tree')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    decision_tree_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(12, 10))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Decision Tree")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Additional Metrics
# -------------------------------
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted')
training_score = decision_tree_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {accuracy_score(y_test, y_pred_test):.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")
#########################
import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.ensemble import GradientBoostingClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from itertools import cycle
from collections import Counter

warnings.filterwarnings('ignore')

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/a.csv")
target_column = "Target"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", Counter(y))

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", Counter(y_res))

# -------------------------------
# 3. Train Gradient Boosting Classifier
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

gb_model = GradientBoostingClassifier(
    n_estimators=100,      # Number of boosting stages
    learning_rate=0.1,     # Step size shrinkage
    max_depth=3,           # Depth of individual trees
    random_state=42
)
gb_model.fit(X_train, y_train)

y_pred_test = gb_model.predict(X_test)
y_score_test = gb_model.predict_proba(X_test)

print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation
# -------------------------------
def explain_model_with_shap(model, X_data, feature_names=None, num_samples=100):
    try:
        explainer = shap.TreeExplainer(model)  # TreeExplainer works well for boosting
        sample = X_data[:num_samples]
        sample_df = pd.DataFrame(sample, columns=feature_names)
        shap_values = explainer.shap_values(sample)
        shap.summary_plot(shap_values, sample_df)
        return shap_values
    except Exception as e:
        print(f"SHAP explanation error: {e}")
        return None

shap_values = explain_model_with_shap(gb_model, X_train, feature_names=X.columns, num_samples=100)

# -------------------------------
# 5. Confusion Matrix (Full Set)
# -------------------------------
data_classes = ['Type 1','Type 2','Type 3c','Gestational','MODY','LADA','Secondary','Neonatal Mellitus',
                'Wolcott-Rallison Syndrome','Steroid-Induced','Cystic Fibrosis-Related','Wolfram Syndrome','Prediabetic']

y_full_pred = gb_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve (Multiclass)
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(12, 10))
colors = cycle(['green', 'orange', 'red', 'blue', 'purple', 'brown', 'pink', 'gray'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - Gradient Boosting')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    gb_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(12, 10))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Gradient Boosting")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Additional Metrics
# -------------------------------
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted')
training_score = gb_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {accuracy_score(y_test, y_pred_test):.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")


#########################3
import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from pyswarm import pso
from itertools import cycle
from collections import Counter

warnings.filterwarnings('ignore')

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/a.csv")
target_column = "Target"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", Counter(y))

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", Counter(y_res))

# -------------------------------
# 3. Train Logistic Regression
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

lr_model = LogisticRegression(max_iter=1000, multi_class='ovr')
lr_model.fit(X_train, y_train)

y_pred_test = lr_model.predict(X_test)
y_score_test = lr_model.predict_proba(X_test)

print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation
# -------------------------------
def explain_model_with_shap(model, X_data, feature_names=None, num_samples=100):
    try:
        explainer = shap.Explainer(model, X_data, check_additivity=False)
        sample = X_data[:num_samples]
        sample_df = pd.DataFrame(sample, columns=feature_names)
        shap_values = explainer(sample)
        shap.summary_plot(shap_values, sample_df)
        return shap_values
    except Exception as e:
        print(f"SHAP explanation error: {e}")
        return None

shap_values = explain_model_with_shap(lr_model, X_train, feature_names=X.columns, num_samples=100)

# -------------------------------
# 5. Confusion Matrix (Full Set)
# -------------------------------
data_classes = ['Type 1','Type 2','Type 3c','Gestational','MODY','LADA','Secondary','Neonatal Mellitus',
                'Wolcott-Rallison Syndrome','Steroid-Induced','Cystic Fibrosis-Related','Wolfram Syndrome','Prediabetic']

y_full_pred = lr_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve (Multiclass)
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(12, 10))
colors = cycle(['green', 'orange', 'red', 'blue', 'purple', 'brown', 'pink', 'gray'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    lr_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(12, 10))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Logistic Regression")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. PSO Hyperparameter Tuning
# -------------------------------
def lr_loss_function(params):
    C = float(params[0])
    clf = LogisticRegression(C=C, max_iter=1000, multi_class='ovr')
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    return 1 - accuracy_score(y_test, preds)

lb = [0.001]
ub = [10.0]
best_params, _ = pso(lr_loss_function, lb, ub, swarmsize=10, maxiter=5)
print("\nBest Hyperparameter C (via PSO):", best_params)

# -------------------------------
# 9. Evaluate Optimized Model
# -------------------------------
opt_lr = LogisticRegression(C=float(best_params[0]), max_iter=1000, multi_class='ovr')
opt_lr.fit(X_train, y_train)
opt_preds = opt_lr.predict(X_test)
opt_proba = opt_lr.predict_proba(X_test)

opt_acc = accuracy_score(y_test, opt_preds)
opt_auc = roc_auc_score(y_test_bin, opt_proba, multi_class='ovr', average='weighted')
opt_f1 = precision_recall_fscore_support(y_test, opt_preds, average='weighted')[2]

comparison = pd.DataFrame({
    "Metric": ["Accuracy", "F1 Score", "AUC"],
    "Before PSO": [
        accuracy_score(y_test, y_pred_test),
        precision_recall_fscore_support(y_test, y_pred_test, average='weighted')[2],
        auc_score
    ],
    "After PSO": [opt_acc, opt_f1, opt_auc]
})

print("\nComparison: Before vs After PSO")
print(comparison.to_string(index=False))

# -------------------------------
# 10. Additional Metrics
# -------------------------------
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted')
training_score = lr_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {accuracy_score(y_test, y_pred_test):.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")
################################
# diabetes_pipeline.py

import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.tree import DecisionTreeClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from pyswarm import pso
from itertools import cycle
from collections import Counter

warnings.filterwarnings('ignore')

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/a.csv")
target_column = "Target"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", Counter(y))

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", Counter(y_res))

# -------------------------------
# 3. Train Decision Tree
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

y_pred_test = dt_model.predict(X_test)
y_score_test = dt_model.predict_proba(X_test)

print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation
# -------------------------------
def explain_model_with_shap(model, X_data, feature_names=None, num_samples=100):
    try:
        explainer = shap.Explainer(model, X_data, check_additivity=False)
        sample = X_data[:num_samples]
        sample_df = pd.DataFrame(sample, columns=feature_names)
        shap_values = explainer(sample)
        shap.summary_plot(shap_values, sample_df)
        return shap_values
    except Exception as e:
        print(f"SHAP explanation error: {e}")
        return None

shap_values = explain_model_with_shap(dt_model, X_train, feature_names=X.columns, num_samples=100)

# -------------------------------
# 5. Confusion Matrix (Full Set)
# -------------------------------
data_classes = ['Type 1','Type 2','Type 3c','Gestational','MODY','LADA','Secondary','Neonatal Mellitus',
                'Wolcott-Rallison Syndrome','Steroid-Induced','Cystic Fibrosis-Related','Wolfram Syndrome','Prediabetic']

y_full_pred = dt_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve (Multiclass)
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(12, 10))
colors = cycle(['green', 'orange', 'red', 'blue', 'purple', 'brown', 'pink', 'gray'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - Decision Tree')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    dt_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(12, 10))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - Decision Tree")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. PSO Hyperparameter Tuning
# -------------------------------
def dt_loss_function(params):
    max_depth = int(params[0])
    clf = DecisionTreeClassifier(max_depth=max_depth, random_state=42)
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    return 1 - accuracy_score(y_test, preds)

lb = [1]
ub = [20]
best_params, _ = pso(dt_loss_function, lb, ub, swarmsize=10, maxiter=5)
print("\nBest Hyperparameter max_depth (via PSO):", best_params)

# -------------------------------
# 9. Evaluate Optimized Model
# -------------------------------
opt_dt = DecisionTreeClassifier(max_depth=int(best_params[0]), random_state=42)
opt_dt.fit(X_train, y_train)
opt_preds = opt_dt.predict(X_test)
opt_proba = opt_dt.predict_proba(X_test)

opt_acc = accuracy_score(y_test, opt_preds)
opt_auc = roc_auc_score(y_test_bin, opt_proba, multi_class='ovr', average='weighted')
opt_f1 = precision_recall_fscore_support(y_test, opt_preds, average='weighted')[2]

comparison = pd.DataFrame({
    "Metric": ["Accuracy", "F1 Score", "AUC"],
    "Before PSO": [
        accuracy_score(y_test, y_pred_test),
        precision_recall_fscore_support(y_test, y_pred_test, average='weighted')[2],
        auc_score
    ],
    "After PSO": [opt_acc, opt_f1, opt_auc]
})

print("\nComparison: Before vs After PSO")
print(comparison.to_string(index=False))

# -------------------------------
# 10. Additional Metrics
# -------------------------------
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted')
training_score = dt_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {accuracy_score(y_test, y_pred_test):.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")
###############################3
# diabetes_pipeline.py

import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.naive_bayes import GaussianNB
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from pyswarm import pso
from itertools import cycle
from collections import Counter

warnings.filterwarnings('ignore')

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/a.csv")
target_column = "Target"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", Counter(y))

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", Counter(y_res))

# -------------------------------
# 3. Train Gaussian Naive Bayes
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)

y_pred_test = gnb_model.predict(X_test)
y_score_test = gnb_model.predict_proba(X_test)

print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation
# -------------------------------
def explain_model_with_shap(model, X_data, feature_names=None, num_samples=100):
    try:
        explainer = shap.Explainer(model.predict_proba, X_data, check_additivity=False)
        sample = X_data[:num_samples]
        sample_df = pd.DataFrame(sample, columns=feature_names)
        shap_values = explainer(sample)
        shap.summary_plot(shap_values, sample_df)
        return shap_values
    except Exception as e:
        print(f"SHAP explanation error: {e}")
        return None

shap_values = explain_model_with_shap(gnb_model, X_train, feature_names=X.columns, num_samples=100)

# -------------------------------
# 5. Confusion Matrix (Full Set)
# -------------------------------
data_classes = ['Type 1','Type 2','Type 3c','Gestational','MODY','LADA','Secondary','Neonatal Mellitus',
                'Wolcott-Rallison Syndrome','Steroid-Induced','Cystic Fibrosis-Related','Wolfram Syndrome','Prediabetic']

y_full_pred = gnb_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve (Multiclass)
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(12, 10))
colors = cycle(['green', 'orange', 'red', 'blue', 'purple', 'brown', 'pink', 'gray'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - GaussianNB')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    gnb_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(12, 10))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - GaussianNB")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Additional Metrics
# -------------------------------
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted')
training_score = gnb_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {accuracy_score(y_test, y_pred_test):.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")
#########################################
# diabetes_pipeline.py

import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.svm import SVC
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from pyswarm import pso
from itertools import cycle
from collections import Counter

warnings.filterwarnings('ignore')

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/a.csv")
target_column = "Target"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", Counter(y))

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", Counter(y_res))

# -------------------------------
# 3. Train Support Vector Machine
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

svm_model = SVC(kernel='rbf', probability=True, random_state=42)
svm_model.fit(X_train, y_train)

y_pred_test = svm_model.predict(X_test)
y_score_test = svm_model.predict_proba(X_test)

print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation
# -------------------------------
def explain_model_with_shap(model, X_data, feature_names=None, num_samples=100):
    try:
        explainer = shap.Explainer(model.predict_proba, X_data, check_additivity=False)
        sample = X_data[:num_samples]
        sample_df = pd.DataFrame(sample, columns=feature_names)
        shap_values = explainer(sample)
        shap.summary_plot(shap_values, sample_df)
        return shap_values
    except Exception as e:
        print(f"SHAP explanation error: {e}")
        return None

shap_values = explain_model_with_shap(svm_model, X_train, feature_names=X.columns, num_samples=100)

# -------------------------------
# 5. Confusion Matrix (Full Set)
# -------------------------------
data_classes = ['Type 1','Type 2','Type 3c','Gestational','MODY','LADA','Secondary','Neonatal Mellitus',
                'Wolcott-Rallison Syndrome','Steroid-Induced','Cystic Fibrosis-Related','Wolfram Syndrome','Prediabetic']

y_full_pred = svm_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve (Multiclass)
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(12, 10))
colors = cycle(['green', 'orange', 'red', 'blue', 'purple', 'brown', 'pink', 'gray'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - SVM')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    svm_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(12, 10))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - SVM")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Additional Metrics
# -------------------------------
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted')
training_score = svm_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {accuracy_score(y_test, y_pred_test):.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")


###########################
# diabetes_pipeline.py

import os
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.model_selection import train_test_split, learning_curve
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, roc_curve, precision_recall_fscore_support,
    mean_squared_error, r2_score
)

from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE, KMeansSMOTE
import shap
from itertools import cycle
from collections import Counter

warnings.filterwarnings('ignore')

# -------------------------------
# 1. Load and preprocess data
# -------------------------------
df = pd.read_csv("c:/a.csv")
target_column = "Target"

features = df.drop(columns=[target_column])
target = df[target_column].astype(int)

# Impute missing values
imputer = IterativeImputer()
features_imputed = imputer.fit_transform(features)

# Standardize features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

X = pd.DataFrame(features_scaled, columns=features.columns)
y = target

# -------------------------------
# 2. Handle class imbalance
# -------------------------------
print("Original class distribution:\n", Counter(y))

kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X, y)

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_resampled, y_resampled)

print("\nResampled class distribution:\n", Counter(y_res))

# -------------------------------
# 3. Train KNN Classifier
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.3, random_state=42, stratify=y_res
)

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

y_pred_test = knn_model.predict(X_test)
y_score_test = knn_model.predict_proba(X_test)

print("\n=== Evaluation on Test Set ===")
print("Accuracy:", accuracy_score(y_test, y_pred_test))
print(classification_report(y_test, y_pred_test))

# -------------------------------
# 4. SHAP Explanation
# -------------------------------
def explain_model_with_shap(model, X_data, feature_names=None, num_samples=100):
    try:
        explainer = shap.Explainer(model.predict_proba, X_data, check_additivity=False)
        sample = X_data[:num_samples]
        sample_df = pd.DataFrame(sample, columns=feature_names)
        shap_values = explainer(sample)
        shap.summary_plot(shap_values, sample_df)
        return shap_values
    except Exception as e:
        print(f"SHAP explanation error: {e}")
        return None

shap_values = explain_model_with_shap(knn_model, X_train, feature_names=X.columns, num_samples=100)

# -------------------------------
# 5. Confusion Matrix (Full Set)
# -------------------------------
data_classes = ['Type 1','Type 2','Type 3c','Gestational','MODY','LADA','Secondary','Neonatal Mellitus',
                'Wolcott-Rallison Syndrome','Steroid-Induced','Cystic Fibrosis-Related','Wolfram Syndrome','Prediabetic']

y_full_pred = knn_model.predict(X_res)
cm_full = confusion_matrix(y_res, y_full_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm_full, annot=True, cmap="YlGnBu", fmt='g')
plt.title("Confusion Matrix - Full Resampled Set")
plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.xticks(np.arange(len(data_classes)), labels=data_classes, rotation=45)
plt.yticks(np.arange(len(data_classes)), labels=data_classes, rotation=0)
plt.tight_layout()
plt.show()

# -------------------------------
# 6. ROC Curve (Multiclass)
# -------------------------------
y_test_bin = label_binarize(y_test, classes=np.unique(y))
n_classes = y_test_bin.shape[1]
fpr, tpr = {}, {}
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score_test[:, i])

plt.figure(figsize=(12, 10))
colors = cycle(['green', 'orange', 'red', 'blue', 'purple', 'brown', 'pink', 'gray'])
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, label=f"{data_classes[i]} vs Rest")
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Multiclass ROC Curve - KNN')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

auc_score = roc_auc_score(y_test_bin, y_score_test, multi_class='ovr', average='weighted')
print(f"Average AUC: {auc_score:.4f}")

# -------------------------------
# 7. Learning Curve
# -------------------------------
train_sizes, train_scores, val_scores = learning_curve(
    knn_model, X_train, y_train, cv=5, scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)

plt.figure(figsize=(12, 10))
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Training')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title("Learning Curve - KNN")
plt.xlabel("Training Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# -------------------------------
# 8. Additional Metrics
# -------------------------------
overall_precision, overall_recall, _, _ = precision_recall_fscore_support(
    y_test, y_pred_test, average='weighted')
training_score = knn_model.score(X_train, y_train)
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print("\n=== Additional Evaluation Metrics ===")
print(f"Overall Accuracy:       {accuracy_score(y_test, y_pred_test):.4f}")
print(f"Overall Precision:      {overall_precision:.4f}")
print(f"Overall Recall:         {overall_recall:.4f}")
print(f"Training Score:         {training_score:.4f}")
print(f"Mean Squared Error:     {mse:.4f}")
print(f"R^2 Score:              {r2:.4f}")
```
