```python
import pandas as pd
import numpy as np
import seaborn as sns
from pyswarm import pso
import matplotlib.pyplot as plt
import shap
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
from sklearn.model_selection import cross_val_score, KFold, train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.preprocessing import StandardScaler
from keras.utils import to_categorical
from imblearn.over_sampling import KMeansSMOTE
from fancyimpute import IterativeImputer
from collections import Counter

warnings.filterwarnings('ignore')
#import pdb; pdb.set_trace()
# -----------------------------
# Load and preprocess data
# -----------------------------
df = pd.read_csv("c:/z310ab.csv")

# Impute missing values using Iterative Imputer (MICE)
imputer = IterativeImputer()
df_imputed = imputer.fit_transform(df)
data = pd.DataFrame(df_imputed, columns=df.columns)

# Separate features and target
y = data["DiagnosisType"].astype(int)
X = data.drop("DiagnosisType", axis=1)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Show original class distribution
print("Original class distribution:", Counter(y))

# -----------------------------
# Apply KMeans-SMOTE
# -----------------------------
kmeans_smote = KMeansSMOTE(random_state=42)
X_resampled, y_resampled = kmeans_smote.fit_resample(X_scaled, y)

# Show new class distribution
print("Resampled class distribution:", Counter(y_resampled))

# -----------------------------
# Model Definitions
# -----------------------------
num_classes = len(np.unique(y_resampled))

def create_nn_model():
    model = Sequential()
    model.add(Dense(12, input_dim=X_resampled.shape[1], activation='relu'))
    model.add(Dense(12, activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

models = {
    "Neural Network": KerasClassifier(build_fn=create_nn_model, epochs=100, batch_size=10, verbose=0),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "SVM": SVC(kernel='rbf', gamma='scale')
}

# -----------------------------
# Cross-Validation
# -----------------------------
results = {}
kf = KFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    scores = cross_val_score(model, X_resampled, y_resampled, cv=kf)
    results[name] = scores
    print(f"{name} Accuracy: {scores.mean():.4f} ± {scores.std():.4f}")

# -----------------------------
# Accuracy Visualization
# -----------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(data=pd.DataFrame(results))
plt.title("Model Accuracy Comparison after KMeans-SMOTE")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()

# -----------------------------
# Train final NN and report
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

final_model = create_nn_model()
final_model.fit(X_train, to_categorical(y_train), epochs=100, verbose=0)
y_pred = np.argmax(final_model.predict(X_test), axis=1)

print("\nFinal Model Evaluation on Test Data:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for Neural Network")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

#----------
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.wrappers.scikit_learn import KerasClassifier

# Convert targets to categorical
num_classes = len(np.unique(y_resampled))
y_res_cat = to_categorical(y_resampled)

# Define model creation function for Keras
def create_nn_model(neurons=12):
    model = Sequential()
    model.add(Dense(int(neurons), input_dim=X_resampled.shape[1], activation='relu'))
    model.add(Dense(int(neurons), activation='relu'))
    model.add(Dense(num_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# Objective function for PSO (NN)
def nn_obj_fn(params):
    neurons, batch_size = int(params[0]), int(params[1])
    model = KerasClassifier(build_fn=lambda: create_nn_model(neurons),
                            epochs=50, batch_size=batch_size, verbose=0)
    scores = cross_val_score(model, X_resampled, y_res_cat.argmax(axis=1), cv=3)
    return -scores.mean()

# PSO bounds: [neurons, batch_size]
nn_bounds = ([10, 8], [100, 64])
nn_best_params, _ = pso(nn_obj_fn, *nn_bounds)
nn_best_params = [int(nn_best_params[0]), int(nn_best_params[1])]
print("Best Neural Network Params:", nn_best_params)

# Train final NN model
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_res_cat, random_state=42)
nn_model = create_nn_model(neurons=nn_best_params[0])
nn_model.fit(X_train, y_train, epochs=50, batch_size=nn_best_params[1], verbose=0)

# Predict and evaluate
y_pred_nn = np.argmax(nn_model.predict(X_test), axis=1)
y_test_labels = np.argmax(y_test, axis=1)

print("\nFinal Neural Network Evaluation:")
print("Accuracy:", accuracy_score(y_test_labels, y_pred_nn))
print("Classification Report:\n", classification_report(y_test_labels, y_pred_nn))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test_labels, y_pred_nn)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Greens")
plt.title("Confusion Matrix - Neural Network")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# SHAP for Neural Network
import shap
explainer = shap.DeepExplainer(nn_model, X_train[:100])
shap_values = explainer.shap_values(X_test[:100])

shap.summary_plot(shap_values, X_test[:100], feature_names=X.columns)
```
